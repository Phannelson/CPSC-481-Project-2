### Axioms:
    Axiom zero:

What logically follows from the knowledge base by comparing what is in the KB with what is resolved from the KB

    Axiom one:

Enforces the rule that if only one call can be true in each row and each column

    Axiom two:

Enforces the rule that if three cells have found to be false within a row or column, then the last cell must be true

    Axiom Three:

Enforces the rule that if two values in two different subgrids are true and they share an element then the value in the third subgrid that is a combination of the first two must be true.
IF THOSE TWO VALUES ARE NOT BOTH TRUE, THEN THE THIRD VALUE MUST BE FALSE.

    Axiom Four:

Although not implemented, the axiom aims to check IF an activity (say hang gliding) is determined to have occurred in a particular year (say 2004) AND a person (say Pam) is determined to have done an activity in that same year (2004), THEN we can infer that the activity done by that person (Pam) in that year (2004) must be that activity (hang gliding).

    Axiom Five:
Although not implemented, the axiom aims to check IF an activity(say hang gliding) occured in a particular year(say 2004) AND  a person(say Pam) did some activity in the same year(2004), THEN it must be that the person(Pam) did not do that particular activity(say hang gliding) in that year(2004).
